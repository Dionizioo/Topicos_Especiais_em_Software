```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold, SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python

# Carregar e combinar os folds de treinamento
train_folds = [pd.read_csv(f'lbp-train-fold_{i}.csv') for i in range(4)]
df_train = pd.concat(train_folds, ignore_index=True)

# Carregar o dataset de teste
df_test = pd.read_csv('lbp-test.csv')

# Verificar a distribuição das classes
print("Distribuição das classes (treinamento):")
print(df_train['class'].value_counts())

print("Distribuição das classes (teste):")
print(df_test['class'].value_counts())
```

```python
# Separar atributos (X) e rótulos (y)
X_train = df_train.drop('class', axis=1)
y_train = df_train['class']
X_test = df_test.drop('class', axis=1)
y_test = df_test['class']
```

```python


# Selecionar atributos com baixa variância
selector_variance = VarianceThreshold(threshold=1e-6)
X_train_reduced = selector_variance.fit_transform(X_train)
X_test_reduced = selector_variance.transform(X_test)

print(f"Número de atributos após seleção por variância: {X_train_reduced.shape[1]}")
```

```python


# Selecionar as 20 melhores features usando ANOVA
selector_anova = SelectKBest(f_classif, k=20)
X_train_anova = selector_anova.fit_transform(X_train_reduced, y_train)
X_test_anova = selector_anova.transform(X_test_reduced)

print(f"Número de atributos após ANOVA: {X_train_anova.shape[1]}")
```

```python


# Balancear classes com SMOTE
smote = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = smote.fit_resample(X_train_anova, y_train)

print("Distribuição das classes após SMOTE:")
print(pd.Series(y_train_balanced).value_counts())
```

```python

# Escalonar os dados
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_balanced)
X_test_scaled = scaler.transform(X_test_anova)
```

```python
# Treinar o modelo
model = RandomForestClassifier(random_state=42)
model.fit(X_train_scaled, y_train_balanced)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Realizar previsões
y_pred = model.predict(X_test_scaled)

# Exibir matriz de confusão
conf_matrix = confusion_matrix(y_test, y_pred)
print("Matriz de Confusão:")
print(conf_matrix)

# Exibir relatório de classificação
print("Relatório de Classificação:")
print(classification_report(y_test, y_pred))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Nomes das classes para o gráfico
class_names = [
    "R/Normal", "COVID-19", "MERS", "SARS", 
    "Varicella", "Streptococcus", "Pneumocystis"
]

# Plotando a Matriz de Confusão
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", 
            xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predito")
plt.ylabel("Verdadeiro")
plt.title("Matriz de Confusão")
plt.show()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# Carregar dados balanceados
X_train = X_train_balanced
y_train = y_train_balanced

# Carregar dados de teste para o fold 0
test_data = pd.read_csv('lbp-train-fold_0.csv')
X_test = test_data.drop(columns=['class'])
y_test = test_data['class']

# Aplicar seleção de variância e ANOVA ao conjunto de teste
X_test_reduced = selector_variance.transform(X_test)
X_test_anova = selector_anova.transform(X_test_reduced)

# Escalonamento dos dados
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test_anova)

# Treinamento do modelo MLP
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)
mlp_model.fit(X_train, y_train)

# Previsões e avaliação
y_pred = mlp_model.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Matriz de confusão
cm = confusion_matrix(y_test, y_pred)
cm_df = pd.DataFrame(cm, index=mlp_model.classes_, columns=mlp_model.classes_)

plt.figure(figsize=(10, 7))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False,
            xticklabels=mlp_model.classes_, yticklabels=mlp_model.classes_)
plt.title('Matriz de Confusão - Rede Neural')
plt.xlabel('Classe Prevista')
plt.ylabel('Classe Real')
plt.show()
```
